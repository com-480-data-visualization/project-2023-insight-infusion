```python
import pandas as pd

df = pd.read_csv("./raw_CO2.csv", sep=";")
df.head()
```

```python
# we just need to translate the food group
x = pd.unique(df["Groupe d'aliment"])
print(len(x))
translated_x = [
    "culinary aids and miscellaneous ingredients", "baby food", "beverages", "starters and mixed dishes", "fruits, vegetables, legumes and oilseeds", 
    "ice creams and sherbets", "milk and dairy products", "fats", "cereal products", "sweetened products", "meat, eggs, fish"
]
print(len(translated_x))
translation_dict = {}
for group_index in range(len(x)): translation_dict[x[group_index]] = translated_x[group_index]

df['Food Group'] = [translation_dict[fr_group] for fr_group in df["Groupe d'aliment"]]
df.head()

```

```python
# we just need to translate the food subgroup
food_subgroup = pd.unique(df["Sous-groupe d'aliment"])
translated_food_subgroup = ['seaweed', 'herbs', 'various ingredients', 'spices', 'cooking aids', 'sauces', 'salts', 'condiments', 'foodstuffs intended for particular nutritional uses', 'infant cereals & cookies', 'infant milks & drinks', 'infant desserts', 'salted snacks & infant dishes', 'alcoholic beverages', 'non-alcoholic beverages', 'water', 'mixed dishes', 'puff pastry & other starters', 'soups', 'vegetarian dishes', 'sandwiches', 'pizzas, pies & savoury pancakes', 'mixed salads & raw vegetables', 'fruits', 'nuts & oil seeds', 'vegetables', 'potatoes & other tubers', 'pulses', 'ice creams', 'frozen desserts', 'sherbets', 'cheeses', 'fresh dairy products & similar', 'creams & cream specialities', 'milks', 'butters', 'vegetable oils & fats', 'other fats', 'fish oils', 'margarines', 'pasta, rice & cereals', 'flours & pastries', 'cakes & pastries', 'breads & pastries', 'breakfast cereals & cookies', 'chocolates & chocolate products', 'non-chocolate confectionery', 'jams & similar', 'sugars, honeys & similar', 'fish & seafood products', 'cooked meat', 'raw meat', 'raw fish', 'cold cuts', 'cooked fish', 'other meat products', 'meat substitutes', 'cooked shellfish', 'raw shellfish', 'eggs', 'cold cut substitutes']

translation_dict = {}
for subgroup_index in range(len(food_subgroup)): translation_dict[food_subgroup[subgroup_index]] = translated_food_subgroup[subgroup_index]

df['Food subgroup'] = [ translation_dict[fr_subgroup] for fr_subgroup in df["Sous-groupe d'aliment"]]
df.tail()
```

```python
import json

CO2_dict = {}

for index, row in df.iterrows():
    CO2_dict[row['LCI Name']] = {
        'agriculture': float(row['Agriculture'].replace(',', '.')),
        'transformation': float(row['Transformation'].replace(',', '.')),
        'packaging': float(row['Emballage'].replace(',', '.')),
        'transportation': float(row['Transport'].replace(',', '.')),
        'distribution': float(row['Supermarché et distribution'].replace(',', '.')),
        'consumption': float(row['Consommation'].replace(',', '.')),
        'total': float(row['Total'].replace(',', '.'))
    }

with open("raw_CO2.json", "w") as write_file:
    json.dump(CO2_dict, write_file,  indent=4)
```

```python
translated_df = df[['LCI Name', 'Agriculture', 'Transformation', 'Emballage', 'Transport', 'Supermarché et distribution', 'Consommation', 'Total', 'Food Group', 'Food subgroup']].copy()
cols = ['Agriculture', 'Transformation', 'Emballage', 'Transport', 'Supermarché et distribution', 'Consommation', 'Total']

for col in cols: translated_df[col] = [float(x.replace(',', '.')) for x in translated_df[col]]

translated_df.head()
```

```python
food_groups = pd.unique(translated_df['Food Group'])
food_subgroups = pd.unique(translated_df['Food subgroup'])

mean_f_groups = {}
for group in food_groups:
    tmp_df = translated_df[translated_df['Food Group'] == group]
    mean_f_groups[group] = tmp_df['Total'].mean()

mean_f_subgroups = {}
for subgroup in food_subgroups:
    tmp_df = translated_df[translated_df['Food subgroup'] == subgroup]
    mean_f_subgroups[subgroup] = tmp_df['Total'].mean()

translated_df['Group mean'] = [mean_f_groups[group] for group in translated_df['Food Group']]
translated_df['Subgroup mean'] = [mean_f_subgroups[subgroup] for subgroup in translated_df['Food subgroup']]

translated_df.head()
```
