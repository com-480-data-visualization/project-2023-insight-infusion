```python
import pandas as pd

df = pd.read_csv("./raw_CO2.csv", sep=";")
df.head()
```

```python
import json

CO2_dict = {}

for index, row in df.iterrows():
    CO2_dict[row['LCI Name']] = {
        'agriculture': row['Agriculture'],
        'transformation': row['Transformation'],
        'packaging': row['Emballage'],
        'transportation': row['Transport'],
        'distribution': row['Supermarché et distribution'],
        'consumption': row['Consommation'],
        'total': row['Total']
    }

with open("raw_CO2.json", "w") as write_file:
    json.dump(CO2_dict, write_file,  indent=4)
```
